```python
def calculate_average(numbers):
    assert isinstance(numbers, list), "Input must be a list"
    assert len(numbers) > 0, "Error: List is empty!"
    return sum(numbers) / len(numbers)

num = [34,46,34,43,434,34,34,23]


calculate_average(num)
```
