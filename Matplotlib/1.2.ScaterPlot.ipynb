# ScaterPlot

```python
import matplotlib.pyplot as plt

hours = [1,2,3,4,5,6,7,8]
scores = [55,60,65,70,75,80,85,90]

plt.scatter(hours,scores, color="green", marker="^", label="Student Data")
plt.xlabel("Hours")
plt.ylabel("Exam score")
plt.title("Relationship b/w hourse and scores")
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.scatter([1,2,3,4],[55,66,77,88], color="blue", marker="^", label="Class A")
plt.scatter([1,2,3],[54,23,90], color="red", marker="^", label="Class B")
plt.xlabel("Hours")
plt.ylabel("Exam score")
plt.title("Relationship b/w hourse and scores")
plt.legend()
plt.grid(True)
plt.show()
```

## **Subplot and Layout Adjesment Function**

Multiple charts on single page or canvas, EDA (explority data analytics). 

```python

x = [1, 2, 4, 5, 6]
y = [16, 24, 42, 15, 36]


plt.subplot(2, 2, 1)  
plt.plot(x, y, marker="o", color="blue")
plt.title("Line Chart")

plt.subplot(2, 2, 2)  
plt.bar(x, y, color="blue", label="y data")
plt.title("Bar Chart")
plt.legend()

plt.subplot(2, 2, 3)  
plt.hist(y, bins=5, color="purple")
plt.title("Histogram")

plt.subplot(2, 2, 4)  
plt.pie(y, labels=x, colors=["purple", "gold", "red", "green", "yellow"], autopct='%1.1f%%')
plt.title("Pie Chart")

plt.tight_layout()
plt.show()
```

```python
# fig, ax = plt.subplot(nrows, ncolumns, figsize=(width, height))

fig, ax = plt.subplots(1,2, figsize=(10,5))

x = [1, 2, 4, 5, 6]
y = [16, 24, 42, 15, 36]

ax[0].plot(x,y, color="blue")
ax[0].set_title("line plot")

ax[1].bar(x,y, color="red")
ax[1].set_title("Bar chart")

fig.suptitle("Charts")

plt.tight_layout()

plt.show()
```
