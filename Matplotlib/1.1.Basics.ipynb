### Matplotlib Functions

```python
months = [1,2,3,4,5]
sales = [100,200,150,300,100]

plt.plot(months, sales, color="blue", linestyle="--", linewidth=2, marker="o", label="2025 sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Data for 2025")
plt.legend(loc="upper left", fontsize=12)
plt.grid(color="gray", linestyle=":", linewidth=1)
plt.show()
```

```python
months = [1,2,3,4,5]
sales = [100,200,150,300,100]

plt.plot(months, sales, color="blue", linestyle="--", linewidth=2, marker="o", label="2025 sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Data for 2025")
plt.legend(loc="upper left", fontsize=12)
plt.grid(color="gray", linestyle=":", linewidth=1)
plt.ylim(0,1000)
plt.xlim(0,10)
plt.xticks(months, ["m1", "m2","m3","m4","m5"])
plt.show()
```

### Pie Charts

```python
region = ["east", "west", "north", "south"]
revenue = [200,344,56,765]

plt.pie(revenue, labels=region, autopct="%1.1f%%", colors=["gold", "blue", "red","lightgreen"] )
plt.title("Revenue by Region")
plt.show()
```

### Histogram

```python
scores = [ 87,56,78,56,54,34,65,76,56,43,23,54, 56, 78, 65, 34, 89, 45, 65, 43, 23, 43]

plt.hist(scores, bins=5, color="purple", edgecolor="black")
plt.ylabel("students")
plt.xlabel("scores range")
plt.show()
```

### Bar Charts

```python
product = ["A", "B", "C", "D", "E"]
sales = [100,200,150,300,100]
plt.bar(product, sales, color="orange", label="Sales 2025")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product and Sales Comparison")
plt.legend()
plt.show()
product = ["A", "B", "C", "D", "E"]
sales = [100,200,150,300,100]
plt.barh(product, sales, color="orange", label="Sales 2025")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product and Sales Comparison")
plt.legend()
plt.show()
```
